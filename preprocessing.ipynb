```python
import numpy as np
from tensorflow.keras.applications.inception_v3 import preprocess_input

def preprocess_frames(frames):
    """
    Preprocess frames for InceptionV3-based models.

    Parameters
    ----------
    frames : np.ndarray
        Shape: (N, H, W, 3)

    Returns
    -------
    np.ndarray
        Preprocessed frames
    """

    frames = preprocess_input(frames)
    return frames


def reshape_for_sequence(frames):
    """
    Ensures frames are correctly shaped for CNN + RNN models.

    Parameters
    ----------
    frames : np.ndarray
        Shape: (N, H, W, 3)

    Returns
    -------
    np.ndarray
        Shape: (1, N, H, W, 3)
    """

    return np.expand_dims(frames, axis=0)
```
